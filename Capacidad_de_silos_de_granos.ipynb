## Planteamiento del Problema:

Se desea calcular la capacidad de almacenamiento de un silo cilíndrico en toneladas de maíz. Para ello, el usuario debe ingresar el radio y la altura del silo en metros. Con estos datos se calcula el volumen del cilindro y posteriormente se multiplica por la densidad del maíz para obtener la cantidad de toneladas que puede almacenar.

##Requerimientos del Problema:
El programa debe solicitar:
El radio del silo (en metros).
La altura del silo (en metros).
Se debe usar la fórmula del volumen del cilindro:


$$V=π⋅r^2⋅h$$

La densidad del maíz es aproximadamente 0.75 ton/m³.
El resultado debe mostrarse en toneladas.
##Prueba del Programa:
Entrada:

$r = 3 m$

$h = 10 m$

$Salida esperada:
≈ 212.06 toneladas$

```python
import math

# Solicitar datos al usuario
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# Densidad del maíz (ton/m^3)
densidad = 0.75

# Cálculo del volumen del cilindro
volumen = math.pi * (radio ** 2) * altura

# Cálculo de la capacidad en toneladas
toneladas = volumen * densidad

# Mostrar resultado
print("La capacidad del silo es:", round(toneladas, 2), "toneladas")
```

import math: Se importa la librería matemática para usar el valor de π (pi).

radio = float(input(...)): Se le pide al usuario que ingrese el radio y se convierte a número decimal.

altura = float(input(...)): Se solicita la altura del silo en metros.

densidad = 0.75: Se define la densidad del maíz en toneladas por metro cúbico.

volumen = math.pi * (radio^2) * altura: Se calcula el volumen del cilindro usando la fórmula.

toneladas = volumen * densidad: Se multiplica el volumen por la densidad para obtener las toneladas.

print(...): Se muestra el resultado final redondeado a 2 decimales.
