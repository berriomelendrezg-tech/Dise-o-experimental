##Planteamiento del Problema:

Se requiere calcular la cantidad de fertilizante líquido y agua necesaria para un sistema de riego por goteo. El usuario debe ingresar el número total de hectáreas y la dosis de fertilizante requerida por hectárea. Con estos datos se determina la cantidad total de fertilizante y la cantidad de agua necesaria según la relación de dilución establecida.

##Requerimientos del Problema:
El programa debe solicitar:

Número de hectáreas.
Dosis de fertilizante (litros por hectárea).

Se debe calcular:

Total de fertilizante:

Fertilizante = hectareas * dosis

Fertilizante = hectareas * dosis

Relación de dilución:

1 litro de fertilizante por 500 litros de agua.

Se debe mostrar:

Litros de fertilizante.

Litros de agua necesarios.
##Prueba del Programa:
Entrada:

Hectáreas = 5

$Dosis = 2 L/ha$

Salida esperada:

10 L de fertilizante

5000 L de agua

```python
# Solicitar datos al usuario
hectareas = float(input("Ingrese el número de hectáreas: "))
dosis = float(input("Ingrese la dosis (litros por hectárea): "))

# Cálculo del fertilizante total
fertilizante = hectareas * dosis

# Relación de dilución (1:500)
agua = fertilizante * 500

# Mostrar resultados
print("Cantidad de fertilizante:", fertilizante, "litros")
print("Cantidad de agua necesaria:", agua, "litros")
```

hectareas = float(input(...)): Se solicita al usuario el número de hectáreas y se convierte a número decimal.

dosis = float(input(...)): Se pide la dosis de fertilizante en litros por hectárea.

fertilizante = hectareas * dosis: Se calcula la cantidad total de fertilizante requerida.

agua = fertilizante * 500: Se calcula la cantidad de agua necesaria usando la relación de dilución 1:500.

print(...): Se muestran los resultados al usuario.
