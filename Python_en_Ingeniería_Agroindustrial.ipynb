## Ejemplos de Código (Sencillos y Útiles)

Aquí tienes tres ejemplos prácticos que podrías aplicar en un entorno real:

### 1. Cálculo del Rendimiento Agrícola

Este código calcula cuánto producto obtienes por unidad de área, una métrica básica para medir la eficiencia del campo.

```
# Datos de entrada
toneladas_cosechadas = 150.5
hectareas_sembradas = 12.0

# Cálculo del rendimiento
rendimiento = toneladas_cosechadas / hectareas_sembradas

print(f"El rendimiento de la cosecha es de {rendimiento:.2f} toneladas por hectárea.")
```

### 2. Monitor de Temperatura para Almacenamiento

Imagina que tienes un sensor en una bodega de refrigeración de frutas. Este programa decide si se debe activar una alerta.

```
temperatura_actual = 8.5  # Dato que vendría de un sensor en tiempo real
limite_maximo = 5.0      # Grados Celsius permitidos para el producto

if temperatura_actual > limite_maximo:
    print("¡ALERTA! La temperatura excede el límite. Revisar sistema de enfriamiento.")
else:
    print("Temperatura dentro del rango óptimo. Producto seguro.")
```

### 3. Balance de Materia (Fórmula de Dilución)

En la industria procesadora, a menudo usamos la fórmula de dilución para ajustar concentraciones de salmueras, almíbares o desinfectantes. La fórmula es $C_1 \times V_1 = C_2 \times V_2$.

```
# Queremos preparar 10 litros (V2) de una solución al 5% (C2)
# partiendo de un concentrado al 20% (C1). ¿Cuánto concentrado (V1) necesito?

c1 = 20.0  # Concentración inicial (%)
c2 = 5.0   # Concentración final deseada (%)
v2 = 10.0  # Volumen final deseado (Litros)

# Despejamos V1: V1 = (C2 * V2) / C1
v1 = (c2 * v2) / c1

print(f"Debes usar {v1} litros del concentrado y completar con agua hasta los {v2} litros.")
```
