##Planteamiento del Problema:

Una planta procesadora de frutas necesita calcular el porcentaje de rendimiento de pulpa obtenida a partir de la fruta procesada. El usuario debe ingresar el peso total de la fruta y el peso de los residuos generados (cáscaras y semillas). Con estos datos se calcula la cantidad de pulpa y el porcentaje de aprovechamiento.

##Requerimientos del Problema:

El programa debe solicitar:

Peso total de la fruta (kg).

Peso de los residuos (kg).

Se debe calcular:

Pulpa:

Pulpa = Peso Total − Residuos

Porcentaje de rendimiento:

%=(Peso Total Pulpa)×100

Se debe mostrar:

Cantidad de pulpa en kg.

Porcentaje de rendimiento.

##Prueba del Programa:
Entrada:

Peso Total = 500 kg

Residuos = 150 kg

Salida esperada:

350 kg de pulpa

70% de rendimiento

```python
# Solicitar datos al usuario
peso_total = float(input("Ingrese el peso total de la fruta (kg): "))
residuos = float(input("Ingrese el peso de los residuos (kg): "))

# Cálculo de la pulpa
pulpa = peso_total - residuos

# Cálculo del porcentaje de rendimiento
rendimiento = (pulpa / peso_total) * 100

# Mostrar resultados
print("Cantidad de pulpa:", pulpa, "kg")
print("Porcentaje de rendimiento:", round(rendimiento, 2), "%")
```

residuos = float(input(...)): Se ingresa el peso de los residuos generados.
