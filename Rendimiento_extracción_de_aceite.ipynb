# **Control de Rendimiento en Extracción de Aceite** #
Este ejercicio es ideal para introducir float, int y la concatenación de strings.

# **Planteamiento:** #
Una planta procesadora de palma de aceite necesita calcular el Índice de Rendimiento de Extracción (IRE). El operario debe ingresar el peso total de los racimos de fruta fresca (RFF) procesados en kilogramos y la cantidad de aceite crudo obtenido en litros.

# **Requerimientos:** #
1. Solicitar el nombre del lote de producción (Tipo string).
2. Solicitar el peso de los racimos en kg (Tipo float). (pr)
3. Solicitar los litros de aceite obtenidos (Tipo float).(lt)
4. Calcular el rendimiento sabiendo que la densidad aproximada del aceite es de 0.92 kg/L. (r)
.

$$ pa (kg)=lt * 0.92 $$
$$ r =(pa/pr) * 100 $$
**Prueba del programa:**

Si se ingresan 1000 kg de racimos y se obtienen 200 litros de aceite, el programa debe informar que el rendimiento del lote es del 18.4%.


````python
# ENTRADA: Solicitar el nombre del lote de producción (Tipo string)
# Solicitar el peso de los racimos en kg (Tipo float). (pr)
# Solicitar los litros de aceite obtenidos (Tipo float).(lt)
NOMBRE = input("Ingrese el nombre del lote: ")
pr = float(input("Ingrese el peso de los racimos en kg: "))
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))

# PROCESO: Calcular el peso del aceite (pa) y el rendimiento (r)
pa = lt * 0.92
r = (pa/pr) * 100

# SALIDA: informar el rendimiento del lote
print("El rendimiento del lote", NOMBRE, "es del", r, "%")
````

A continuación, se explica cada línea del código proporcionado para el cálculo del Índice de Rendimiento de Extracción (IRE):

```python
NOMBRE = input("Ingrese el nombre del lote: ")
```
Esta línea solicita al usuario que ingrese el nombre del lote de producción. La función `input()` muestra el mensaje entre paréntesis al usuario y espera a que este escriba algo y presione Enter. El texto ingresado se guarda como una cadena de caracteres (string) en la variable `NOMBRE`.

```python
pr = float(input("Ingrese el peso de los racimos en kg: "))
```
Esta línea pide al usuario que ingrese el peso total de los racimos de fruta fresca (RFF) en kilogramos. Similar a la anterior, `input()` obtiene el texto del usuario. La función `float()` convierte ese texto a un número decimal (tipo float), lo cual es necesario para operaciones matemáticas, y lo almacena en la variable `pr`.

```python
lt = float(input("Ingrese la cantidad de aceite crudo obtenido en litros: "))
```
De manera análoga a la línea anterior, esta línea solicita la cantidad de aceite crudo obtenida en litros. El valor ingresado se convierte a un número decimal (`float`) y se guarda en la variable `lt`.

```python
pa = lt * 0.92
```
Aquí se calcula el peso del aceite obtenido en kilogramos (`pa`). Para ello, se multiplica la cantidad de aceite en litros (`lt`) por su densidad aproximada, que es 0.92 kg/L.

```python
r = (pa/pr) * 100
```
Esta línea calcula el rendimiento de extracción (`r`). Primero, se divide el peso del aceite (`pa`) entre el peso de los racimos (`pr`). El resultado de esta división se multiplica por 100 para expresarlo como un porcentaje.

```python
print("El rendimiento del lote", NOMBRE, "es del", r, "%")
```
Finalmente, esta línea muestra el resultado al usuario. Utiliza la función `print()` para concatenar varias partes: un texto literal ("El rendimiento del lote"), el nombre del lote ingresado por el usuario (`NOMBRE`), otro texto literal ("es del"), el valor calculado del rendimiento (`r`), y el símbolo de porcentaje ("%"). Esto forma un mensaje completo que informa el rendimiento del lote.
