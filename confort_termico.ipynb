# **Metodos y calculos**

 ### **Índice de temperatura y humedad (producción intensiva)**

 $$𝐼𝑇𝐻 = 0.72(𝑇_(𝑏𝑠) + 𝑇𝑏ℎ) + 40.6 $$

 donde:

 1. $T_{bs}$: temperatura bulbo seco ($^oc$)
 2. $T_{bn}$: temperatura de bulbo humedo ($^oC$)

 Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

+ si $ITH <= 70$ ambiente no estresante
+ si $ITH > 70andITH$ <= 78$ situacion alerta por estres termico
+ si $ITH > 78andITH$ <= 83$ peligro por estres termico
+ si $ITH > 83$ si tuacion critica

```python
 # ENTRADA: Temperatura de bulbo
 #Temperatura de bulbo humedo (tbs), grados centigrados
tbs= float(input("ingrese la temperatura de bulbo seco,grados centigrados: "))
tbh= float(input("ingrese la temperatura de bulbo humedo,grados centigrados: "))

# PROCESO: Calculo de ith
ith= 0.72 * (tbs + tbh) + 40.6

#SALIDA: ith
if ith <= 70:
  print("ambiente no estresante")
elif ith > 70 and ith <= 78:
  print("situacion alerta por estres termico")
elif ith > 78 and ith <= 83:
  print("peligro por estres termico")
elif ith  > 83:
  print("si tuacion critica")

```
